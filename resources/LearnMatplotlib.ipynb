```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of plots

### Line plots

```python
x = [1, 2, 3, 4, 5]
y =  [10, 20, 30, 35, 40]
```

```python
plt.plot(x, y);
plt.title("Simple Line Plot");
plt.xlabel("X-axis Label");
plt.ylabel("Y-axis Label");
plt.grid(True)
```

### Bar Chart

```python
months = ["Jan", "Feb", "Mar", "Apr"]
revenue = [100, 230, 400, 620]
```

```python
plt.figure(figsize=(10, 6));
plt.bar(months, revenue);
plt.title("Monthly Revenue Bar Chart");
plt.xlabel("Months");
plt.ylabel("Revenue");
```

### Scatter Plot

```python
heights = [65, 70, 67, 71]
weights = [150, 160, 156, 170]
```

```python
plt.scatter(heights, weights);
plt.xlabel("Heights (inches)");
plt.ylabel("Weights (pounds)");
plt.title("Heights vs. Weights Scatter Plot");
```

### Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, "b-o", label="Original Price");
plt.plot(days, predicted_price,"r-o", label="Predicted Price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Price");
```

### Subplots

```python
# First subplot - Original price
plt.subplot(3, 1, 1);
plt.plot(days, original_price, "b-o", label="Original Price");
plt.title("Original Price Over time");
plt.xlabel("Weekdays");
plt.ylabel("Price");
plt.legend();

# Second subplot - Predicted price
plt.subplot(3, 1, 2);
plt.plot(days, predicted_price,"r-o", label="Predicted Price");
plt.title("Predicted Price Over time");
plt.xlabel("Weekdays");
plt.ylabel("Price");
plt.legend();


# Third subplot - Difference between Original & Predicted Prices
price_diff = [0 - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3);
plt.bar(days, price_diff, color="purple", label="Price Price");
plt.title("Difference between Original & Predicted Prices");
plt.xlabel("Days");
plt.ylabel("Price Difference");
plt.legend();


plt.tight_layout()
```

### Saving the Figure

```python
plt.plot(days, original_price, "b-o", label="Original Price");
plt.plot(days, predicted_price,"r-o", label="Predicted Price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Price");
plt.savefig("example_prediction.png", dpi=300) # Save the image
```
